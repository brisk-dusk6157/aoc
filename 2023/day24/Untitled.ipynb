```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, cos, sin
from sympy.plotting import plot3d_parametric_line
%matplotlib ipympl
```

```python
arr = np.load('arr.npy')
```

```python
print(arr.shape)
print(arr)
```

```python
t = symbols('t')
x0, y0, z0 = 24, 13, 10
dx0, dy0, dz0 = -3, 1, 2
rock = (x0 + dx0 * t, y0 + dy0 * t, z0 + dz0 * t, (t, 0, 6))
plot3d_parametric_line(rock, *[(x + dx*t, y + dy*t, z + dz * t, (t, 0, 6)) for x, y, z, dx, dy, dz in arr])
```

```python
arr = np.load('arr_full.npy')

t = symbols('t')
tmax = 500000000000

plot3d_parametric_line(*[(x + dx*t, y + dy*t, z + dz * t, (t, 0, tmax)) for x, y, z, dx, dy, dz in arr])
```
